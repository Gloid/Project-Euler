```python
# Problem 3
upper_bound = round(600851475143**0.5) + 1
num = 600851475143

prime_list = [2, 3]
divisors = []

prime_candidate = 3

while num > 1:
    t = prime_list[-1]

    if num % t == 0:
        divisors.append(t)
        while num % t == 0:
            num = num // t

    prime_candidate = prime_candidate + 2
    found_prime = False

    while not found_prime:
        is_prime = True
        for prime in prime_list:
            if prime_candidate % prime == 0:
                is_prime = False
                break

        if is_prime:
            prime_list.append(prime_candidate)
            found_prime = True
        else:
            prime_candidate += 2

print(max(divisors))
```

```python
# Problem 4

def is_palindrome(num):
    num = str(num)
    for i, digit1 in enumerate(num, start=1):
        digit2 = num[-i]
        if digit1 != digit2:
            return False
    return True
palindrome_list = []
for x in range(100, 1000):
    for y in range(x, 1000):
        product = x*y
        if is_palindrome(product):
            palindrome_list.append(product)
print(max(palindrome_list))
```
